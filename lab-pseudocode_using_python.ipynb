### Exercise 2: Check Even or Odd

Write pseudocode that checks if a number is even or odd and prints the result. Then, write the equivalent Python code.

<!-- Add your pseudocode here -->

```python
# pseudocode:
# BEGIN
#  DEFINE function check_even_odd(number)
#    IF number MOD2 = 0 THEN
#       PRINT number is "Even"
#    ELSE
#       PRINT number is "Odd"
#  END FUNCTION

#  READ number from user input: 
#   TRY to CONVERT input to integer
#       CALL check_even_odd(number)
#       Print result
#   HANDLE invalid input
#       Print "Error: Please enter a valid number!" if input is invalid
#   END

# python code to check even or odd: (using function)

def check_even_odd(number):
    """Check if a number is even or odd and return the result as a string."""
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"
    
try:
    # Read a number from user input
    number = int(input("Enter a number: "))
    
    # Call the function and get the result
    result = check_even_odd(number)
    
    # Print the result
    print(f"The number {number} is {result}.")
except ValueError:
    print("Error: Please enter a valid integer!")
    
```

### Exercise 3: Sum of First N Natural Numbers

Write pseudocode to calculate the sum of the first N natural numbers. Then, write the equivalent Python code.

<!-- Add your pseudocode here -->

```python
# pseudocode
# BEGIN
#    START TRY BLOCK
#        DISPLAY "Enter a positive integer N: "
#        READ input_value
#        SET N = CONVERT_TO_INTEGER(input_value)
        
#        IF N < 0 THEN
#            RAISE ValueError WITH MESSAGE "N must be a positive integer!"
#        END IF
        
#        SET sum_N = (N ร (N + 1)) รท 2
#        DISPLAY "The sum of the first " + N + " natural numbers is: " + sum_N
        
#    CATCH (ValueError exception)
#        IF exception_message = "N must be a positive integer!" THEN
#            DISPLAY "Error: N must be a positive integer!"
#        ELSE
#            DISPLAY "Error: Please enter a valid integer!"
#        END IF
#    END TRY-CATCH
# END
```

```python
# pythoon code (using mathematical expression directly):
# mathematical formula:
#   Sum of first N natural numbers = N * (N + 1) / 2

# Python code (with proper error handling):
try:
    # Read N from user input
    N = int(input("Enter a positive integer N: "))
    
    if N < 0:
        raise ValueError("N must be a positive integer!")
    
    # Calculate the sum of the first N natural numbers
    sum_N = N * (N + 1) // 2
    
    # Display the result
    print(f"The sum of the first {N} natural numbers is: {sum_N}")
    
except ValueError as e:
    if str(e) == "N must be a positive integer!":
        print("Error: N must be a positive integer!")
    else:
        print("Error: Please enter a valid integer!")
```

### Exercise 5: Count Vowels in a String

Write pseudocode to count the number of vowels in a given string and print the count. Then, write the equivalent Python code.

<!-- Add your pseudocode here -->

```python
# pseudocode
# text
# BEGIN
#    READ input_string
#    SET vowel_count = 0
#    SET vowels = "aeiouAEIOU"
    
#    FOR each character IN input_string DO
#        IF character IS IN vowels THEN
#            SET vowel_count = vowel_count + 1
#        END IF
#    END FOR
    
#    PRINT "Number of vowels:", vowel_count
# END

# python
# Read a string from user input
input_string = input("Enter a string/word/sentence: ")

# Initialize vowel count
vowel_count = 0
vowels = "aeiouAEIOU"

# Count vowels in the string
for char in input_string:
    if char.lower() in vowels:
        vowel_count += 1

# Display the result
print(f"Number of vowels: {vowel_count}")
```

### Bonus Exercise 6: Reverse a String

Write pseudocode to reverse a given string and print it out.

<!-- Add your pseudocode here -->

```python
# pseudocode
# BEGIN
#    DISPLAY "Enter a string: "
#    READ input_string
    
#    SET reversed_string = EMPTY_STRING
#    SET string_length = LENGTH(input_string)
#    SET index = string_length - 1
    
#    WHILE index >= 0 DO
#        SET current_char = input_string[index]
#        SET reversed_string = reversed_string + current_char
#        SET index = index - 1
#    END WHILE
    
#    DISPLAY "Original: " + input_string
#    DISPLAY "Reversed: " + reversed_string
# END

# python
# Read input string
input_string = input("Enter a string/word/sentence: ")

# Initialize variables
reversed_string = ""
string_length = len(input_string)
index = string_length - 1

# Loop backwards using while loop
while index >= 0:
    current_char = input_string[index]
    reversed_string = reversed_string + current_char
    index = index - 1

# Print results
print("Original:", input_string)
print("Reversed:", reversed_string)
```

### Bonus Exercise 7: Print Multiplication Table of a Number

Write pseudocode to print the multiplication table of a given number up to 10.

<!-- Add your pseudocode here -->

```python
# pseudocode
# BEGIN
#    DISPLAY "Enter a number: "
#    READ number
    
#    DISPLAY "Multiplication table for", number, ":"
    
#    SET multiplier = 1
#    REPEAT
#        SET product = number * multiplier
#        DISPLAY number, " ร ", multiplier, " = ", product
#        SET multiplier = multiplier + 1
#    UNTIL multiplier > 10
# END

# python
# DISPLAY "Enter a number: "
number = int(input("Enter a number: "))

# DISPLAY "Multiplication table for", number, ":"
print(f"Multiplication table for {number}:")

# SET multiplier = 1
multiplier = 1

# REPEAT
while True:
    # SET product = number * multiplier
    product = number * multiplier
    
    # DISPLAY number, " ร ", multiplier, " = ", product
    print(f"{number} ร {multiplier} = {product}")
    
    # SET multiplier = multiplier + 1
    multiplier = multiplier + 1
    
    # UNTIL multiplier > 10
    if multiplier > 10:
        break
```

### Bonus Exercise 9: Find Factorial of a Number

Write pseudocode that calculates the factorial of a given positive integer.

<!-- Add your pseudocode here -->

```python
# pseudocode
# BEGIN
#    DISPLAY "Enter a positive integer: "
#    READ number
#    SET factorial = 1
    
#    IF number < 0 THEN
#        PRINT "Error: Factorial is not defined for negative numbers"
#    ELSE IF number = 0 THEN
#        PRINT "Factorial of 0 is 1"
#    ELSE
#       SET i = 1
#        WHILE i <= number DO
#            SET factorial = factorial * i
#            SET i = i + 1
#        END WHILE
#        PRINT "Factorial of", number, "is", factorial
#    END IF
# END

# python
# Read number from user input
number = int(input("Enter a positive integer: "))

# Initialize factorial
factorial = 1

# Check for valid input
if number < 0:
    print("Error: Factorial is not defined for negative numbers")
elif number == 0:
    print("Factorial of 0 is 1")
else:
    # Calculate factorial using while loop
    i = 1
    while i <= number:
        factorial = factorial * i
        i = i + 1
    
    print(f"Factorial of {number} is {factorial}")
```

### Bonus Exercise 10: FizzBuzz Problem

Write pseudocode for the FizzBuzz problem where you print numbers from 1 to N but replace multiples of three with "Fizz" and multiples of five with "Buzz".

<!-- Add your pseudocode here -->

```python
# pseudocode
# BEGIN
#    DISPLAY "Enter the value of N: "
#    READ N
    
#    FOR i FROM 1 TO N DO
#        SET output = EMPTY_STRING
        
#        IF i MOD 3 = 0 THEN
#            SET output = output + "Fizz"
#        END IF
        
#        IF i MOD 5 = 0 THEN
#            SET output = output + "Buzz"
#        END IF
        
#        IF output = EMPTY_STRING THEN
#            SET output = CONVERT_TO_STRING(i)
#        END IF
        
#        DISPLAY output
#    END FOR
# END

# python:
# Read N from user input
N = int(input("Enter the value of N: "))

# FizzBuzz using string concatenation
for i in range(1, N + 1):
    output = ""
    
    if i % 3 == 0:
        output += "Fizz"
    
    if i % 5 == 0:
        output += "Buzz"
    
    # If output is still empty, use the number
    if output == "":
        output = str(i)
    
    print(output)
```
